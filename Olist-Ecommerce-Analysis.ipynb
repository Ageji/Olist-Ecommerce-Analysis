```python
us_finance.columns = us_finance.columns.str.strip()
```

```python
us_finance.drop(columns=["Multiyear Working Capital Extension"], inplace=True)
```

```python
us_finance.isnull().sum().sort_values(ascending=False)
```

```python
us_finance.replace(["NA","N/A",""],np.nan,inplace=True)
```

```python
us_finance.isnull().sum().sort_values(ascending=False)
```

```python
print(us_finance.duplicated(subset=['Unique Identifier']).sum())
```

```python
us_finance.head()
```

```python
date_cols=['Decision Date','Effective Date','Expiration Date']
for cols in date_cols:
    us_finance[cols]=pd.to_datetime(us_finance[cols],errors='coerce')
```

```python
us_finance.info()
```

```python
print(us_finance['Loan Interest Rate'].value_counts())
```

```python
us_finance['Loan Interest Rate(Numeric)']=pd.to_numeric(us_finance['Loan Interest Rate'],errors='coerce')
us_finance['Loan Interest Rate(Category)']=us_finance['Loan Interest Rate'].where(
    us_finance['Loan Interest Rate(Numeric)'].isna() & us_finance['Loan Interest Rate'].notna()
)
us_finance['Loan Interest Rate(Category)']=us_finance['Loan Interest Rate(Category)'].replace('Undisbursed',np.nan)
us_finance=us_finance.drop(columns=['Loan Interest Rate'])
```
